## One side

```python
def get_onesided_spectrum(E):
    idxp = np.where(E.freq_time.values>0)[0]
    idxn = np.where(E.freq_time.values<0)[0]
    omega_p = E.freq_time.values[idxp]
    E_p = E.isel(freq_time=idxp)
    E_n = E.isel(freq_time=idxn)
    E_n = E_n.assign_coords(freq_time=-E_n.freq_time).sortby('freq_time')
    return E_n + E_p
```

```python
def dask_argmax(ratio_log10,omega):
    # r shape = 1 x 140/190 x 120/190 x 1440
    # omega 1440
    #return omega[np.argmax(r<1,axis=-1)]*0 + omega.shape[1]
    #return omega[np.argmax(ratio_log10<0.5, axis=-1)]
    idx = ratio_log10.shape[-1] - np.argmax(ratio_log10[...,::-1]<threshold, axis=-1) - 1
    return omega[idx]
#return np.argwhere(ratio_log10<0.5)[-1]
#return (omega[None,None,:]+0.*r)[np.argmin(r,axis=-1)]
#return omega
#return np.argmin(r,axis=-1)
```
