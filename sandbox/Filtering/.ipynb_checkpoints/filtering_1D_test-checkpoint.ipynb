# SSU



```python
v = 'SSU'

## open files, nc files
ncdata_dir = '/home1/scratch/xyu/mit_nc/'

# dir where filtered data is stored
out_dir = '/home1/scratch/xyu/mit_filtered/'
```

```python
face = 5

# search for right files, for example: SSU_f01_t*.nc
filenames = sorted(glob(ncdata_dir+v+'_f%02d_t*.nc' %(face)))
print('Number of files available: %d' %len(filenames))
#filenames = filenames[1:5] # tmpÂ£
print('Number of files processed: %d' %len(filenames))

grid = xr.open_dataset(scratch+'/mit_grid_t.nc')
grid = grid.isel(face=face).drop(['XG','YG','i_g','j_g','k','k_u','k_l','k_p1'])
print(grid)
```

```python
%time ds0 = xr.open_mfdataset(filenames, chunks={'i':10}, concat_dim='time', compat='equals')

ds = ds0

ds = ds.assign_coords(time_days=ds['time']/86400.)

print(ds)
print('\n data size: %.1f GB' %(ds.nbytes / 1e9))
```

________
# Temporal filtering (1D along time Dimensions) 

```python
lds = ds.isel(i=0, j=0)
wt = lds['SSU'].window

cutoff = 1.5 # A 1.5-day cutoff
dx = 1./24. # Define the sampling period (one day)
wt.set(n=24*1, dim='time', cutoff=cutoff, dx=dx, window='boxcar')
wt.plot()
```

```python
res_valid = wt.convolve(trim=True)
```

```python
plt.figure(figsize=(16,6))
ax=plt.subplot(111)

toplt =  lds['SSU'].isel(time=slice(0,24*100)).swap_dims({'time': 'time_days'})
toplt.plot(ax=ax)
#
toplt =  res_valid.isel(time=slice(0,24*100)).swap_dims({'time': 'time_days'})
toplt.plot(ax=ax)
#
toplt =  (lds['SSU']-res_valid-2.).isel(time=slice(0,24*100)).swap_dims({'time': 'time_days'})
toplt.plot(ax=ax)

ax.set_xlim(60., 180.)
ax.set_ylim(-1., 1.)
ax.grid()
```

# Spatial filtering (1D along time dimension)

```python
lds = ds
wt = lds['SSU'].window
print(wt)
```

```python
cutoff = 1.5 # A 1.5-day cutoff
dx = 1./24. # Define the sampling period (one day)
wt.set(n=24*1, dim='time', cutoff=cutoff, dx=dx, window='boxcar')
print(wt)
```

```python
lds_f = wt.convolve(trim=True)
print(lds)
```

```python
plt.figure(figsize=(16,12))

it=100

eta =  ds0['SSU'].isel(time=it, i=slice(2500,4320), j=slice(0,2000)) # 432 file read, because of i chunks
#eta =  lds['Eta'].isel(time=it) # note: xarray open all files in order to load Eta here because of the large chunk along time
eta = eta.persist()

etaf = (lds['SSU']-lds_f).isel(time=it, i=slice(2500,4320), j=slice(0,2000)) # 1836 file read
etaf = etaf.persist()

ax=plt.subplot(221)
eta.plot(ax=ax, x='i', y='j', vmin=-1, vmax=1, cmap = cm.balance)
ax.set_title('SSU')

#
ax=plt.subplot(222)
etaf.plot(ax=ax, x='i', y='j', vmin=-1, vmax=1, cmap = cm.balance)
ax.set_title('low pass filtered SSU')

#
ax=plt.subplot(223)
(eta-etaf).plot(ax=ax, x='i', y='j', vmin=-1, vmax=1, cmap = cm.balance)
ax.set_title('high pass filtered SSU')
```

# Save files

_____

```python
# store in netcdf files

Nt = 24*10 # time windows to consider
fsize_bound = 1e20
it = np.arange(lds.time.size/Nt-1).astype(int)*Nt
#it = np.arange(10).astype(int)*Nt # tmp
assert it[-1]+Nt<lds.time.size
#
#for face in range(ds['face'].size):
for i, t in enumerate(it):
    #
    file_out = out_dir+'/%s_filtered_f%02d_t%02d.nc'%(v, face, i)
    if not os.path.isfile(file_out) or os.path.getsize(file_out) < fsize_bound:            
        dv = lds_f.isel(time=slice(t,t+Nt))
        #
        dv = dv.chunk({'time': dv['time'].size, 'i': 432, 'j': 432})
        chunksizes = [432, 432, dv['time'].size]
        #
        dv = dv.load() # this may be required in order to prevent pickle related errors
        #
        while True:
            try:
                %time dv.to_netcdf(file_out, mode='w', unlimited_dims=['time'], \
                                   encoding={'SST': {'chunksizes': chunksizes}})
            except:
                print('Failure')
            break
            #if os.path.isfile(file_out) and os.path.getsize(file_out)>fsize_bound:
            #    #
            #    print('face=%d / i=%d'%(face,i))
            #    break
    else:
        print('face=%d / i=%d - allready processed'%(face,i))
```
