# 1. Read dataset


```python
raw_dir_work = '/home1/datawork/xyu/OSMOSIS_llc4320/data_rechunk/'

grid = xr.open_dataset(raw_dir_work+'Grid.nc', decode_coords=True)
ds_U = xr.open_zarr(raw_dir_work+'U_total_rechunk.zarr')
ds_V = xr.open_zarr(raw_dir_work+'V_total_rechunk.zarr')
ds_W = xr.open_zarr(raw_dir_work+'W_total_rechunk.zarr')
ds_B = xr.open_zarr(raw_dir_work+'Buoy_total_rechunk.zarr')

ds = xr.merge([ds_U,ds_V, ds_W, ds_B, grid])
ds = generate_grid_ds(ds, {'Z':'depth'})
print(ds)
print('\n data size: %.1f GB' %(ds.nbytes / 1e9))
```

# 2. Filter

## low-pass filter over 1 day

```python
out_dir_zarr = '/home1/scratch/xyu/KE/data_rechunk_filter/'

def butter_filter(x, b, a):
    return xr.apply_ufunc(partial(filtfilt, b, a),
                          x.chunk(),
                          dask='parallelized',
                          output_dtypes=[x.dtype],
                          kwargs={'axis': 0})
```

```python
lowcut = 1/24/3600

fs = 1/3600
nyq = 0.5*fs
low = lowcut/nyq

b, a = butter(4, low, 'low')
```

```python
# rechunk data first
#chunks = {'time': 9415, 'lat': 1, 'lon_g': 1, 'depth':52}
#U = ds.U.chunk(chunks)
#del U.depth.encoding['chunks']

# filter
U_filter = butter_filter(ds.U, b, a)
U_low_filter = np.real(U_filter).rename('U_low_filter')
U_low_filter = U_low_filter.to_dataset()

# rechunk data again
chunks = {'time': 9415, 'lat': 1, 'lon_g': 177, 'depth':1}
U_low_filter = U_low_filter.chunk(chunks)
print(U_low_filter)

# save
file_out = out_dir_zarr+'/U_1day_lowpass_filter.zarr'
%time U_low_filter.to_zarr(file_out, mode='w')   
```

```python
# rechunk data first
#chunks = {'time': 9415, 'lat_g': 1, 'lon': 1, 'depth':52}
#V = ds.V.chunk(chunks)
#del V.depth.encoding['chunks']

# filter
V_filter = butter_filter(ds.V, b, a)
V_low_filter = np.real(V_filter).rename('V_low_filter')
V_low_filter = V_low_filter.to_dataset()

# rechunk data again
chunks = {'time': 9415, 'lat_g': 1, 'lon': 177, 'depth':1}
V_low_filter = V_low_filter.chunk(chunks)
print(V_low_filter)

# save
file_out = out_dir_zarr+'/V_1day_lowpass_filter.zarr'
%time V_low_filter.to_zarr(file_out, mode='w')   
```

```python
# rechunk data first
#chunks = {'time': 9415, 'lat': 1, 'lon': 1, 'depth':52}
#W = ds.W.chunk(chunks)
#del W.depth.encoding['chunks']

# filter
W = ds.W
W_filter = butter_filter(W, b, a)
W_low_filter = np.real(W_filter).rename('W_low_filter')
W_low_filter = W_low_filter.to_dataset()

# rechunk data again
chunks = {'time': 9415, 'lat': 1, 'lon': 177, 'depth':1}
W_low_filter = W_low_filter.chunk(chunks)
print(W_low_filter)
# save
file_out = out_dir_zarr+'/W_1day_lowpass_filter.zarr'
%time W_low_filter.to_zarr(file_out, mode='w')   
```

```python
# rechunk data first
#chunks = {'time': 9415, 'lat': 1, 'lon': 1, 'depth':52}
#B = ds.Buoy.chunk(chunks)
#del B.depth.encoding['chunks']

# filter
B_filter = butter_filter(ds.Buoy, b, a)
B_low_filter = np.real(B_filter).rename('B_low_filter')
B_low_filter = B_low_filter.to_dataset()

# rechunk data again
chunks = {'time': 9415, 'lat': 1, 'lon': 177, 'depth':1}
B_low_filter = B_low_filter.chunk(chunks)
print(B_low_filter)
# save
file_out = out_dir_zarr+'/B_1day_lowpass_filter.zarr'
%time B_low_filter.to_zarr(file_out, mode='w')   
```
