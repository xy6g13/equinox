## Read AVISO data

```python
out_dir_zarr = '/home1/datawork/xyu/OSMOSIS_llc4320/data_store/'
ds = xr.open_zarr(out_dir_zarr+'AVISO_uv_total_rechunk.zarr')
print(ds)
print('\n data size: %.1f GB' %(ds.nbytes / 1e9))
```

```python
ds_full = generate_grid_ds(ds, {'X':'lon', 'Y':'lat'})
ds_full
```

```python
grid = Grid(ds_full, periodic=['X','Y'])
grid
```

```python
dlong = grid.diff(ds_full.lon, 'X', boundary_discontinuity=360)
dlonc = grid.diff(ds_full.lon_left, 'X', boundary_discontinuity=360)
dlonc_wo_discontinuity = grid.diff(ds_full.lon_left, 'X')

dlatg = grid.diff(ds_full.lat, 'Y', boundary='fill', fill_value=np.nan)
dlatc = grid.diff(ds_full.lat_left, 'Y', boundary='fill', fill_value=np.nan)
dlatg[0]=0.25
dlatc[-1]=0.25
```

```python
def dll_dist(dlon, dlat, lon, lat):
        """Converts lat/lon differentials into distances in meters

        PARAMETERS
        ----------
        dlon : xarray.DataArray longitude differentials
        dlat : xarray.DataArray latitude differentials
        lon  : xarray.DataArray longitude values
        lat  : xarray.DataArray latitude values

        RETURNS
        -------
        dx  : xarray.DataArray distance inferred from dlon
        dy  : xarray.DataArray distance inferred from dlat
        """

        distance_1deg_equator = 111000.0
        
        dx = dlon * xr.ufuncs.cos(xr.ufuncs.deg2rad(lat)) * distance_1deg_equator
        dy = ((lon * 0) + 1) * dlat * distance_1deg_equator
        return dx, dy
```

```python
ds_full.coords['dxg'], ds_full.coords['dyg'] = dll_dist(dlong, dlatg, ds_full.lon, ds_full.lat)
ds_full.coords['dxc'], ds_full.coords['dyc'] = dll_dist(dlonc, dlatc, ds_full.lon, ds_full.lat)
ds_full.coords['area_c'] = ds_full.dxc * ds_full.dyc
ds_full
```

```python
ds_full.dxg.plot(x='lon_left',y='lat')
```

```python
ds_full.dyg.plot(x='lon',y='lat_left')
```

```python
du_dx = grid.diff(ds_full.u.where(ds.u>-20), 'X') / ds_full.dxg
du_dx_ij = grid.interp(du_dx, 'X')
du_dx_ij = np.real(du_dx_ij).rename('du_dx_ij') 
du_dx_ij
```

```python
du_dy = grid.diff(ds_full.u.where(ds.u>-20), 'Y') / ds_full.dyg
du_dy_ij = grid.interp(du_dy, 'Y')
du_dy_ij = np.real(du_dy_ij).rename('du_dy_ij') 
du_dy_ij
```

```python
dv_dx = grid.diff(ds_full.v.where(ds.u>-20), 'X') / ds_full.dxg
dv_dx_ij = grid.interp(dv_dx, 'X')
dv_dx_ij = np.real(dv_dx_ij).rename('dv_dx_ij') 
dv_dx_ij
```

```python
dv_dy = grid.diff(ds_full.v.where(ds.u>-20), 'Y') / ds_full.dyg
dv_dy_ij = grid.interp(dv_dy, 'Y')
dv_dy_ij = np.real(dv_dy_ij).rename('dv_dy_ij') 
dv_dy_ij
```

```python
def iters_to_date(iters, delta_t=24*3600.):
    t0 = datetime(2012,1,1)    
    ltime = delta_t * (np.array(iters))
    dtime = [t0+dateutil.relativedelta.relativedelta(seconds=t) for t in ltime]    
    return dtime
```

```python
time_day = iters_to_date(np.arange(731))
```

```python
ds_cg = xr.merge([ds_full, du_dx_ij, du_dy_ij, dv_dx_ij, dv_dy_ij])
ds_cg
```

```python
ds_cg = ds_cg.assign_coords(time=time_day) 
print(ds_cg)
```

# Spatial filter

```python
out_dir_zarr = '/home1/datawork/xyu/OSMOSIS_llc4320/data_store/'
ds_cg = xr.open_zarr(out_dir_zarr+'AVISO_uv_rechunk_spatial.zarr')
print(ds_cg)
print('\n data size: %.1f GB' %(ds_cg.nbytes / 1e9))
```

```python
grid_num = 4  # 4*2km = 8km
#grid_num = 10 # 10*2km = 20km
#grid_num = 15 # 15*2km = 30km
#grid_num = 25 # 25*2km = 50km
```

```python
wt_U = ds_cg['u'].window
wt_U.set(n={'lat': grid_num, 'lon': grid_num}, window={'lat':'hanning', 'lon':'hanning'}, chunks = {'lat': 720, 'lon': 1440, 'time':1})
ds_U_filter = wt_U.convolve()

wt_V = ds_cg['v'].window
wt_V.set(n={'lat': grid_num, 'lon': grid_num}, window={'lat':'hanning', 'lon':'hanning'}, chunks = {'lat': 720, 'lon': 1440, 'time':1})
ds_V_filter = wt_V.convolve()

wt_U2 = (ds_cg['u']**2).window
wt_U2.set(n={'lat': grid_num, 'lon': grid_num}, window={'lat':'hanning', 'lon':'hanning'}, chunks = {'lat': 720, 'lon': 1440, 'time':1})
ds_U2_filter = wt_U2.convolve()

wt_V2 = (ds_cg['v']**2).window
wt_V2.set(n={'lat': grid_num, 'lon': grid_num}, window={'lat':'hanning', 'lon':'hanning'}, chunks = {'lat': 720, 'lon': 1440, 'time':1})
ds_V2_filter = wt_V2.convolve()

uv = ds_cg.v*ds_cg.u
wt_UV = uv.window
wt_UV.set(n={'lat': grid_num, 'lon': grid_num}, window={'lat':'hanning', 'lon':'hanning'}, chunks = {'lat': 720, 'lon': 1440, 'time':1})
ds_UV_filter = wt_UV.convolve()

wt_du_dx = ds_cg['du_dx_ij'].window
wt_du_dx.set(n={'lat': grid_num, 'lon': grid_num}, window={'lat':'hanning', 'lon':'hanning'}, chunks = {'lat': 720, 'lon': 1440, 'time':1})
du_dx_filter = wt_du_dx.convolve()

wt_du_dy = ds_cg['du_dy_ij'].window
wt_du_dy.set(n={'lat': grid_num, 'lon': grid_num}, window={'lat':'hanning', 'lon':'hanning'}, chunks = {'lat': 720, 'lon': 1440, 'time':1})
du_dy_filter = wt_du_dy.convolve()

wt_dv_dx = ds_cg['dv_dx_ij'].window
wt_dv_dx.set(n={'lat': grid_num, 'lon': grid_num}, window={'lat':'hanning', 'lon':'hanning'}, chunks = {'lat': 720, 'lon': 1440, 'time':1})
dv_dx_filter = wt_dv_dx.convolve()

wt_dv_dy = ds_cg['dv_dy_ij'].window
wt_dv_dy.set(n={'lat': grid_num, 'lon': grid_num}, window={'lat':'hanning', 'lon':'hanning'}, chunks = {'lat': 720, 'lon': 1440, 'time':1})
dv_dy_filter = wt_dv_dy.convolve()
```

```python
# f stands for low-pass filter

PI_1 = (ds_U2_filter-ds_U_filter**2)*du_dx_filter  # [f(u^2) - f(u)^2]*f(du/dx)
PI_2 = (ds_V2_filter-ds_V_filter**2)*dv_dy_filter  # [f(v^2) - f(v)^2]*f(dv/dy)
PI_3 = (ds_UV_filter - ds_U_filter*ds_V_filter )*(du_dy_filter+dv_dx_filter) # [f(uv) - f(u)f(v)]*[f(du/dy) + f(dv/dx)]

PI_cg = -(PI_1+PI_2+PI_3)
PI_cg
```

```python
PI_cg.persist()
```

```python
fig = plt.figure(figsize=(36,18))
ax = plt.subplot(111)
PI_cg.sel(time=slice('2012-09-05', '2013-09-05')).where(ds_cg.u>-20).mean(['time']).plot(x='lon',y='lat',vmin=-1e-8,vmax=1e-8,cmap=cm.balance)
ax.scatter(343.8125,48.6875,s=32,marker='x',color='r')
```
