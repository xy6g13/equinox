```python
def quick_llc_plot(data, axis_off=False, **kwargs):
    face_to_axis = {0: (2, 0), 1: (1, 0), 2: (0, 0),
                    3: (2, 1), 4: (1, 1), 5: (0, 1),
                    7: (0, 2), 8: (1, 2), 9: (2, 2),
                    10: (0, 3), 11: (1, 3), 12: (2, 3)}
    transpose = [7, 8, 9, 10, 11, 12]
    gridspec_kw = dict(left=0, bottom=0, right=4, top=4, wspace=0, hspace=0)
    fig, axes = plt.subplots(nrows=3, ncols=4, gridspec_kw=gridspec_kw)
    for face, (j, i) in face_to_axis.items():
        data_ax = data.isel(face=face)
        ax = axes[j,i]
        yincrease = True
        if face in transpose:
            data_ax = data_ax.transpose()
            yincrease = False
        data_ax.plot(ax=ax, yincrease=yincrease, **kwargs)
        if axis_off:
            ax.axis('off')
        ax.set_title('')
```

# 2 Linear Coriolis terms

## G_u(fv) ---> SSV
Next we create a `Grid` object from the dataset.
We need to tell xgcm that the `X` and `Y` axes are periodic.
(The other axes will be assumed to be non-periodic.)

```python
#import xgcm
grid = xgcm.Grid(ds, periodic=['X', 'Y'])
grid
```

```python
SSV_dxG = ds.dxG * ds['SSV']
print(SSV_dxG)
```

```python
dxgSSV_j = grid.interp(SSV_dxG,'Y')
print(dxgSSV_j)
```

```python
dxgSSV_ji = grid.interp(dxgSSV_j,'X')
print(dxgSSV_ji)
```

```python
dxgSSV_ji_time1 = dxgSSV_ji.isel(time=1)
print(dxgSSV_ji_time1)
```

```python
lat = ds['YG']
print(lat)
```

```python
omega = 7.3/100000
f = 2*omega*np.sin(np.deg2rad(lat))
print(f)
```

```python
f_j = grid.interp(f,'Y')
print(f_j)
```

```python
Coriolis_u = f_j * dxgSSV_ji_time1 /ds.dxC
print(Coriolis_u)
```

```python
Coriolis_u_time1_face1 = Coriolis_u.isel(face=1)
print(Coriolis_u_time1_face1)
```

```python
Coriolis_u_time1_face1.plot(vmin=-0.0001, vmax=0.0001, cmap = cm.balance)
```

# 3. Time derivative of SSU

A central difference is applied to compute du/dt.

```python
SSU_time2 = ds['SSU'].isel(time=2)
print(SSU_time2)
```

```python
SSU_time0 = ds['SSU'].isel(time=0)
print(SSU_time0)
```

```python
d_SSU_dt_time1 = (SSU_time2 - SSU_time0)/2/3600
print(d_SSU_dt_time1)
```

```python
d_SSU_dt_time1_face1 = d_SSU_dt_time1.isel(face=1)
print(d_SSU_dt_time1_face1)
```

```python
d_SSU_dt_time1_face1.plot(vmin=-0.0001, vmax=0.0001, cmap = cm.balance)
```

# 4. Relative vorticity

Here we compute the vertical component of the vorticity from the velocity field. It is defined as

$$ \zeta = - \frac{\partial u}{\partial y} + \frac{\partial v}{\partial x} \ . $$

On the c-grid, a finite-volume representation is given by

$$ \zeta =  (- \delta_j \Delta x_c u + \delta_i \Delta y_c v ) / A_\zeta \ . $$

In xgcm, we calculate this quanity at (i_g, j_g)

```python
SSV_dyC = ds.dyC * ds['SSV']
d_SSV_dyC_di = grid.diff(SSV_dyC,'X')
print(d_SSV_dyC_di)
```

```python
SSU_dxC = ds.dxC * ds['SSU']
d_SSU_dxC_dj = grid.diff(SSU_dxC,'Y')
print(d_SSU_dxC_dj)
```

```python
zeta_3 = (d_SSV_dyC_di - d_SSU_dxC_dj)/ds.rAz
print(zeta_3)
```

```python
zeta_3_time1 = zeta_3.isel(time=1)
print(zeta_3_time1)
```

```python
zeta_3_time1_j = grid.interp(zeta_3_time1,'Y')
print(zeta_3_time1_j)
```

```python
Coriolis_zeta3_u = zeta_3_time1_j * dxgSSV_ji_time1 /ds.dxC
print(Coriolis_zeta3_u)
```

```python
Coriolis_zeta3_u_time1_face1 = Coriolis_zeta3_u.isel(face=1)
```

```python
(Coriolis_zeta3_u_time1_face1).plot(vmin=-0.0001, vmax=0.0001, cmap = cm.balance)
```

# 5. KE
Finally, we calculate the kinetic energy $1/2 (u^2 + v^2)$ by interpolating both quantities into the cell center point (i.e. (i, j)).

```python
SSU_2 = ds['SSU'] * ds['SSU']
SSU_2_i = grid.interp(SSU_2,'X')
print(SSU_2_i)
```

```python
SSV_2 = ds['SSV'] * ds['SSV']
SSV_2_j = grid.interp(SSV_2,'Y')
print(SSV_2_j)
```

```python
KE = (SSU_2_i + SSV_2_j)/2
print(KE)
```

```python
dKE_dx = (x_grid.diff(KE,'X'))/ds.dxC
print(dKE_dx)
```

```python
dKE_dx_time1_face1 = dKE_dx.isel(time=1,face=1)
print(dKE_dx_time1_face1)
```

```python
dKE_dx_time1_face1.plot(vmin=-0.0001, vmax=0.0001, cmap = cm.balance)
```

# 6. Residuals

residuals_u = du/dt - coriolis_linear - coriolis_nonliear +  gradient_of_Bernoulli_function

```python
residuals_u_time1_face1 = d_SSU_dt_time1_face1 - Coriolis_u_time1_face1 - Coriolis_zeta3_u_time1_face1 + dphi_dx_time1_face1 + dKE_dx_time1_face1
print(residuals_u_time1_face1)
```
