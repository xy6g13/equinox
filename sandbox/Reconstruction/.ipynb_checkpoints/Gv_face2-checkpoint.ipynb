# Read dataset


```python
Nf = 720
data_dir = '/home1/scratch/xyu/mit_nc_t/'

filenames_Eta = sorted(glob(data_dir+'Eta_*.nc'))
filenames_Eta = filenames_Eta[:Nf]
ds_Eta = xr.open_mfdataset(filenames_Eta, chunks={'i':432, 'j':432}, concat_dim='time', compat='equals')


filenames_SSU = sorted(glob(data_dir+'SSU_*.nc'))
filenames_SSU = filenames_SSU[:Nf]
ds_SSU = xr.open_mfdataset(filenames_SSU, chunks={'i_g':432, 'j':432}, concat_dim='time', compat='equals')
#chunks={'time': len(ds_SSU['time'])}
#ds_SSU = ds_SSU.chunk(chunks)  
    
filenames_SSV = sorted(glob(data_dir+'SSV_*.nc'))
filenames_SSV = filenames_SSV[:Nf]
ds_SSV = xr.open_mfdataset(filenames_SSV, chunks={'i':432, 'j_g':432}, concat_dim='time', compat='equals')
#chunks={'time': len(ds_SSV['time'])}
#ds_SSV = ds_SSV.chunk(chunks)  

grid = xr.open_dataset('/home1/scratch/xyu/mit_grid.nc', decode_coords=True)
```

```python
ds = xr.merge([ds_Eta, ds_SSU, ds_SSV, grid])
print(ds)
```

```python
time_begin = 1
time_end = Nf-1
```

# 1. Gradient of Bernoulli function (Eta)

# 1.1 face connections

```python
face_connections = {'face': {0: {'X': ((12, 'Y', False), (3, 'X', False)),
                                 'Y': (None, (1, 'Y', False))},
                             1: {'X': ((11, 'Y', False), (4, 'X', False)),
                                 'Y': ((0, 'Y', False), (2, 'Y', False))},
                             2: {'X': ((10, 'Y', False), (5, 'X', False)),
                                 'Y': ((1, 'Y', False), (6, 'X', False))},
                             3: {'X': ((0, 'X', False), (9, 'Y', False)),
                                 'Y': (None, (4, 'Y', False))},
                             4: {'X': ((1, 'X', False), (8, 'Y', False)),
                                 'Y': ((3, 'Y', False), (5, 'Y', False))},
                             5: {'X': ((2, 'X', False), (7, 'Y', False)),
                                 'Y': ((4, 'Y', False), (6, 'Y', False))},
                             6: {'X': ((2, 'Y', False), (7, 'X', False)),
                                 'Y': ((5, 'Y', False), (10, 'X', False))},
                             7: {'X': ((6, 'X', False), (8, 'X', False)),
                                 'Y': ((5, 'X', False), (10, 'Y', False))},
                             8: {'X': ((7, 'X', False), (9, 'X', False)),
                                 'Y': ((4, 'X', False), (11, 'Y', False))},
                             9: {'X': ((8, 'X', False), None),
                                 'Y': ((3, 'X', False), (12, 'Y', False))},
                             10: {'X': ((6, 'Y', False), (11, 'X', False)),
                                  'Y': ((7, 'Y', False), (2, 'X', False))},
                             11: {'X': ((10, 'X', False), (12, 'X', False)),
                                  'Y': ((8, 'Y', False), (1, 'X', False))},
                             12: {'X': ((11, 'X', False), None),
                                  'Y': ((9, 'Y', False), (0, 'X', False))}}}
```

```python
x_grid = xgcm.Grid(ds, face_connections=face_connections)
print(x_grid)
```

# 1.2 Gradient of 'pressure' (i.e. dphi/dy)

```python
g = 9.81
```

```python
dphi_dy = (x_grid.diff(g*ds['Eta'],'Y', boundary='extend'))/ds.dyC

print(dphi_dy)
```

```python
dphi_dy_face2 = dphi_dy.isel(face=2,time=slice(time_begin,time_end)).load()
print(dphi_dy_face2)
```

_______

# 2 Linear Coriolis terms

## 2.1 G_u(fv) ---> SSV

```python
#import xgcm
grid = xgcm.Grid(ds, periodic=['X', 'Y'])
grid
```

```python
SSU_dyG = ds.dyG * ds['SSU']
print(SSU_dyG)
```

```python
dyGSSU_i = grid.interp(SSU_dyG,'X')
print(dyGSSU_i)
```

```python
dyGSSU_ij = grid.interp(dyGSSU_i,'Y')
print(dyGSSU_ij)
```

```python
lat = ds['YG']
print(lat)
```

```python
omega = 7.3/100000
f = 2*omega*np.sin(np.deg2rad(lat))
print(f)
```

```python
f_i = grid.interp(f,'X')
print(f_i)
```

```python
Coriolis_v = -f_i * dyGSSU_ij /ds.dyC
print(Coriolis_v)   
```

```python
Coriolis_v_face2 = Coriolis_v.isel(face=2,time=slice(time_begin,time_end)).load()
print(Coriolis_v_face2)
```

# 3 Time derivative of SSV

```python
SSV_forward = ds['SSV'].isel(time=slice(time_begin+1,time_end+1))
SSV_forward = SSV_forward.drop(['dtime','time'])
print(SSV_forward)
```

```python
SSV_backward = ds['SSV'].isel(time=slice(time_begin-1,time_end-1))
SSV_backward = SSV_backward.drop(['dtime','time'])
print(SSV_backward)
```

```python
d_SSV_dt_center = (SSV_forward - SSV_backward)/2/3600
print(d_SSV_dt_center)
```

```python
d_SSV_dt_center_face2 = d_SSV_dt_center.isel(face=2).load()
print(d_SSV_dt_center_face2)
```

# 4. Relative vorticity

```python
SSV_dyC = ds.dyC * ds['SSV']
d_SSV_dyC_di = grid.diff(SSV_dyC,'X')
print(d_SSV_dyC_di)
```

```python
SSU_dxC = ds.dxC * ds['SSU']
d_SSU_dxC_dj = grid.diff(SSU_dxC,'Y')
print(d_SSU_dxC_dj)
```

```python
zeta_3 = (d_SSV_dyC_di - d_SSU_dxC_dj)/ds.rAz
print(zeta_3)
```

```python
zeta_3_i = grid.interp(zeta_3,'X')
print(zeta_3_i)
```

```python
Coriolis_zeta3_v = -zeta_3_i * dyGSSU_ij /ds.dyC
print(Coriolis_zeta3_v)
```

```python
Coriolis_zeta3_v_face2 = Coriolis_zeta3_v.isel(face=2,time=slice(time_begin,time_end)).load()
print(Coriolis_zeta3_v_face2)
```

# 5. KE

```python
SSU_2 = ds['SSU'] * ds['SSU']
SSU_2_i = grid.interp(SSU_2,'X')
print(SSU_2_i)
```

```python
SSV_2 = ds['SSV'] * ds['SSV']
SSV_2_j = grid.interp(SSV_2,'Y')
print(SSV_2_j)
```

```python
KE = (SSU_2_i + SSV_2_j)/2
print(KE)
```

```python
dKE_dy = (x_grid.diff(KE,'Y', boundary='extend'))/ds.dyC
print(dKE_dy)
```

```python
dKE_dy_face2 = dKE_dy.isel(time=slice(time_begin,time_end),face=2).load()
print(dKE_dy_face2)
```

# 6. Residuals

```python
residuals_v_face2 = (d_SSV_dt_center_face2 - Coriolis_v_face2 - Coriolis_zeta3_v_face2 + dphi_dy_face2 + dKE_dy_face2).load()
print(residuals_v_face2)

# dir where spectra will be stored
#E_dir = '/home1/scratch/xyu/'

#file_E = E_dir+'residuals_v_face2.nc'   
#residuals_v_face2.to_netcdf(file_E, mode='w')  
```
