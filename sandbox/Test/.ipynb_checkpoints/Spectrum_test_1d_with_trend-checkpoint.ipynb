```python
import numpy as np
import xarray as xr
from matplotlib import pyplot as plt
%matplotlib inline

import xrft
from glob import glob
```

```python
ds = xr.Dataset(coords={'time': ('time', np.arange(3000)*3600.)})
ds = ds.assign(z=np.cos(2.*np.pi*2/86400.*ds.time))
ds = ds.assign(u=ds.z+np.random.randn(ds.time.size))
ds = ds.assign(v=ds.u+np.arange(3000)/1000)

print(ds)
```

```python
fig, ax = plt.subplots(figsize=(20,8))

ax = plt.subplot(3,1,1)
ds.z.plot()
# Label, size
ax.set_xlabel('Time',fontsize=16)
ax.set_ylabel('u',fontsize=16)
ax.tick_params(labelsize=16)

ax = plt.subplot(3,1,2)
ds.u.plot()
# Label, size
ax.set_xlabel('Time',fontsize=16)
ax.set_ylabel('u',fontsize=16)
ax.tick_params(labelsize=16)

ax = plt.subplot(3,1,3)
ds.v.plot()
# Label, size
ax.set_xlabel('Time',fontsize=16)
ax.set_ylabel('v',fontsize=16)
ax.tick_params(labelsize=16)
```

# 2. Normal FFT (i.e. one segment, no overlapping)

## 2.1 applying a hanning window

```python
u_normal = ds['u']

u_fft_normal = xrft.dft(u_normal.fillna(0.), shift=False, detrend='linear', window=False)
E_u_normal_w = (u_fft_normal*u_fft_normal.conj()).where(u_fft_normal!=0., drop=False)/3000*3600
print(E_u_normal_w)
E_u_normal_w = np.real(E_u_normal_w).rename('u_power_spectrum')
```

```python
v_normal = ds['v']

v_fft_normal = xrft.dft(v_normal.fillna(0.), shift=False, detrend=False, window=False)
E_v_normal_w = (v_fft_normal*v_fft_normal.conj()).where(v_fft_normal!=0., drop=False)/3000*3600
print(E_v_normal_w)
E_v_normal_w = np.real(E_v_normal_w).rename('v_power_spectrum')
```

```python
# diff_f
df=E_u_normal_w.freq_time[1]-E_u_normal_w.freq_time[0]
print(df)
```

```python
E_u_normal_w = E_u_normal_w.assign_coords(freq_time=E_u_normal_w.freq_time * 86400.) 
E_u_normal_w
```

```python
E_v_normal_w = E_v_normal_w.assign_coords(freq_time=E_v_normal_w.freq_time * 86400.) 
E_v_normal_w
```

```python
fig, ax = plt.subplots(figsize=(24,8))

labels = ['u', 'v']

(E_u_normal_w/8.64).where(E_u_normal_w.freq_time>0).plot(ax=ax,label=labels[0])
(E_v_normal_w/8.64).where(E_v_normal_w.freq_time>0).plot(ax=ax,label=labels[1])

ax.legend(); ax.grid()
ax.set_xscale('log'); 
ax.set_yscale('log')
ax.set_ylabel(''); 

ax.set_ylim([1e-1,1e6])
ax.set_xlim([8*1e-3,1*1e1])
```

```python
(E_u_normal_w.where((abs(E_u_normal_w.freq_time)>1.99) & (abs(E_u_normal_w.freq_time)<2.01) , other=0.)*df).sum(dim='freq_time').compute() #####
```

```python
(E_v_normal_w.where((abs(E_v_normal_w.freq_time)>1.99) & (abs(E_v_normal_w.freq_time)<2.10) , other=0.)*df).sum(dim='freq_time').compute() #####
```
