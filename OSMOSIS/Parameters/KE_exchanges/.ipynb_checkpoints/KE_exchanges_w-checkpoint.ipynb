# 1. Read dataset


```python
out_dir_zarr = '/home1/scratch/xyu/OSMOSIS_llc4320/zarr_file/'

grid = xr.open_dataset('/home1/datawork/xyu/OSMOSIS_llc4320/data_rechunk/Grid.nc', decode_coords=True)
ds_U_bandpass = xr.open_zarr(out_dir_zarr+'U_bandpass_filter_rechunk.zarr')
ds_V_bandpass = xr.open_zarr(out_dir_zarr+'V_bandpass_filter_rechunk.zarr')
ds_W_bandpass = xr.open_zarr(out_dir_zarr+'W_bandpass_filter_rechunk.zarr')

ds_U_lowpass = xr.open_zarr(out_dir_zarr+'U_lowpass_filter_rechunk.zarr')
ds_V_lowpass = xr.open_zarr(out_dir_zarr+'V_lowpass_filter_rechunk.zarr')
ds_W_lowpass = xr.open_zarr(out_dir_zarr+'W_lowpass_filter_rechunk.zarr')

ds_filter = xr.merge([ds_U_bandpass,ds_V_bandpass,ds_W_bandpass,ds_U_lowpass,ds_V_lowpass,ds_W_lowpass,grid])
ds_filter = generate_grid_ds(ds_filter, {'Z':'depth'})

print(ds_filter)
```

```python
gridx = xgcm.Grid(ds_filter, coords={'X':{'center':'lon', 'left':'lon_g'},
                                     'Y':{'center':'lat', 'left':'lat_g'},
                                     'Z':{'center':'depth', 'left':'depth_left'}}, periodic=['X', 'Y'])
print(gridx)
```

```python
# Calculate vertical distances located on the cellboundary
ds_filter.coords['dzc'] = gridx.diff(ds_filter.depth, 'Z', boundary='extrapolate')
# Calculate vertical distances located on the cellcenter
ds_filter.coords['dzt'] = gridx.diff(ds_filter.depth_left, 'Z', boundary='extrapolate')
ds_filter
```

# 2. Calculate PI_omega (only terms involved with w)

```python
u_prime = gridx.interp(ds_filter.U_bandpass_filter, 'X')
u_prime
```

```python
v_prime = gridx.interp(ds_filter.V_bandpass_filter, 'Y')
v_prime
```

```python
w_prime = ds_filter.W_bandpass_filter.drop(['CS','SN','XC','YC','dzt','rA','Depth'])
print(w_prime)
```

```python
dW_dx_1 = gridx.diff(ds_filter.W_low_filter, 'X')/ds_filter.dxC
dW_dx =  gridx.interp(dW_dx_1,'X')
dW_dx
```

```python
dW_dy_1 = gridx.diff(ds_filter.W_low_filter, 'Y')/ds_filter.dyC
dW_dy =  gridx.interp(dW_dy_1,'Y')
dW_dy
```

### 2.1 vertical shears

```python
dU_dz_left = -1*gridx.diff(ds_filter.U_low_filter, 'Z', boundary='fill', fill_value=np.nan) / ds_filter.dzc
dU_dz_1 = gridx.interp(dU_dz_left,'Z', boundary='fill')
dU_dz = gridx.interp(dU_dz_1,'X')
dU_dz = dU_dz.astype(np.float32)
print(dU_dz)
```

```python
dV_dz_left = -1*gridx.diff(ds_filter.V_low_filter, 'Z', boundary='fill', fill_value=np.nan) / ds_filter.dzc
dV_dz_1 = gridx.interp(dV_dz_left,'Z', boundary='fill')
dV_dz = gridx.interp(dV_dz_1,'Y')
dV_dz = dV_dz.astype(np.float32)
print(dV_dz)
```

```python
dW_dz_left = -1*gridx.diff(ds_filter.W_low_filter, 'Z', boundary='fill', fill_value=np.nan) / ds_filter.dzc
dW_dz = gridx.interp(dW_dz_left,'Z', boundary='fill')
dW_dz = dW_dz.astype(np.float32)
print(dW_dz)
```

## 2.2 store

```python
Pi_w_1 = -(u_prime*w_prime*dW_dx)
chunks = {'time': 9415, 'lat': 201, 'lon': 177, 'depth':1}
Pi_w_1 = Pi_w_1.chunk(chunks)
Pi_w_1 = np.real(Pi_w_1).rename('Pi_w_1')
Pi_w_1 = Pi_w_1.to_dataset()
print(Pi_w_1)
file_out = out_dir_zarr+'/Pi_vertical_rechunk_1.zarr'
%time Pi_w_1.to_zarr(file_out, mode='w') 
```

```python
Pi_w_2 = -(v_prime*w_prime*dW_dy)
chunks = {'time': 9415, 'lat': 201, 'lon': 177, 'depth':1}
Pi_w_2 = Pi_w_2.chunk(chunks)
Pi_w_2 = np.real(Pi_w_2).rename('Pi_w_2')
Pi_w_2 = Pi_w_2.to_dataset()
print(Pi_w_2)
file_out = out_dir_zarr+'/Pi_vertical_rechunk_2.zarr'
%time Pi_w_2.to_zarr(file_out, mode='w') 
```

```python
Pi_w_3 = -(w_prime*w_prime*dW_dz)
chunks = {'time': 9415, 'lat': 201, 'lon': 177, 'depth':1}
Pi_w_3 = Pi_w_3.chunk(chunks)
Pi_w_3 = np.real(Pi_w_3).rename('Pi_w_3')
Pi_w_3 = Pi_w_3.to_dataset()
print(Pi_w_3)  
file_out = out_dir_zarr+'/Pi_vertical_rechunk_3.zarr'
%time Pi_w_3.to_zarr(file_out, mode='w') 
```

```python
Pi_w_4 = -(u_prime*w_prime*dU_dz)
chunks = {'time': 9415, 'lat': 201, 'lon': 177, 'depth':1}
Pi_w_4 = Pi_w_4.chunk(chunks)
Pi_w_4 = np.real(Pi_w_4).rename('Pi_w_4')
Pi_w_4 = Pi_w_4.to_dataset()
print(Pi_w_4)  
file_out = out_dir_zarr+'/Pi_vertical_rechunk_4.zarr'
%time Pi_w_4.to_zarr(file_out, mode='w') 
```

```python
Pi_w_5 = -(v_prime*w_prime*dV_dz)
chunks = {'time': 9415, 'lat': 201, 'lon': 177, 'depth':1}
Pi_w_5 = Pi_w_5.chunk(chunks)
Pi_w_5 = np.real(Pi_w_5).rename('Pi_w_5')
Pi_w_5 = Pi_w_5.to_dataset()
print(Pi_w_5)  
file_out = out_dir_zarr+'/Pi_vertical_rechunk_5.zarr'
%time Pi_w_5.to_zarr(file_out, mode='w') 
```

## 2.3 sum to get PI_omega (w)

```python
Pi_w_1 = xr.open_zarr(out_dir_zarr+'Pi_vertical_rechunk_1.zarr')
Pi_w_2 = xr.open_zarr(out_dir_zarr+'Pi_vertical_rechunk_2.zarr')
Pi_w_3 = xr.open_zarr(out_dir_zarr+'Pi_vertical_rechunk_3.zarr')
Pi_w_4 = xr.open_zarr(out_dir_zarr+'Pi_vertical_rechunk_4.zarr')
Pi_w_5 = xr.open_zarr(out_dir_zarr+'Pi_vertical_rechunk_5.zarr')
```

```python
Pi_w = Pi_w_1.Pi_w_1+Pi_w_2.Pi_w_2+Pi_w_3.Pi_w_3+Pi_w_4.Pi_w_4+Pi_w_5.Pi_w_5
chunks = {'time': 9415, 'lat': 201, 'lon': 177, 'depth':1}
Pi_w = Pi_w.chunk(chunks)
Pi_w = np.real(Pi_w).rename('Pi_w')
Pi_w = Pi_w.to_dataset()
print(Pi_w)
file_out = out_dir_zarr+'/Pi_vertical_rechunk.zarr'
%time Pi_w.to_zarr(file_out, mode='w')   
```
