```python
out_dir_zarr = '/home/c11-data/Test_aponte/xiaolong/Parameter/LLC4320/filtered_uv_data/'
ds_KE = xr.open_zarr(out_dir_zarr+'KE_mean.zarr')
ds_KE
```

```python
out_dir_store = '/home/c11-data/Test_aponte/xiaolong/Parameter/LLC4320/gaussian/'
PI_cg = xr.open_zarr(out_dir_store+'PI_cg.zarr')
PI_cg = PI_cg.PI_cg

out_dir = '/home1/datawork/xyu/OSMOSIS_llc4320/data_rechunk/'
ds_MLD = xr.open_zarr(out_dir+'mld_filter.zarr')

ds = xr.merge([PI_cg,ds_MLD])
# define (real) time
def iters_to_date(iters, delta_t=3600.):
    t0 = datetime.datetime(2011,9,13,0)    
    ltime = delta_t * (np.array(iters))
    dtime = [t0+dateutil.relativedelta.relativedelta(seconds=t) for t in ltime]    
    return dtime

time_day = iters_to_date(np.arange(9415))

ds = ds.assign_coords(time=time_day) 
ds.PI_cg.persist()
ds
```

```python
PI_cg_smean = ds.PI_cg.isel(lat=slice(1,-1),lon=slice(1,-1)).mean(['lon','lat'])
PI_cg_cumsum = 3600*ds.PI_cg.isel(lat=slice(1,-1),lon=slice(1,-1)).mean(['lon','lat']).cumsum(['time'])
PI_cg_smean.load()
PI_cg_cumsum.load()
```
