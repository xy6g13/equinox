# 1. Read dataset



```python
out_dir_work = '/home1/datawork/xyu/OSMOSIS_llc4320/data_rechunk/'


grid = xr.open_dataset(out_dir_work+'Grid.nc', decode_coords=True)
ds_U = xr.open_zarr(out_dir_work+'U_total_rechunk.zarr')
ds_V = xr.open_zarr(out_dir_work+'V_total_rechunk.zarr')

ds = xr.merge([ds_U,ds_V,grid])
ds = generate_grid_ds(ds, {'Z':'depth'})

# define (real) time
def iters_to_date(iters, delta_t=3600.):
    t0 = datetime.datetime(2011,9,13,0)    
    ltime = delta_t * (np.array(iters))
    dtime = [t0+dateutil.relativedelta.relativedelta(seconds=t) for t in ltime]    
    return dtime

time_day = iters_to_date(np.arange(9415))
ds = ds.assign_coords(time=time_day) 
print(ds)
print('\n data size: %.1f GB' %(ds.nbytes / 1e9))
```

```python
U = ds.U.isel(depth=16) - ds.U.isel(depth=16).mean(['time'])
V = ds.V.isel(depth=16) - ds.V.isel(depth=16).mean(['time'])
```

```python
EKE = (gridx.interp(U**2,'X') + gridx.interp(V**2,'Y'))/2 # KE
EKE
```

# 2. read filtered data

```python
out_dir_zarr = '/home/c11-data/Test_aponte/xiaolong/Parameter/LLC4320/filtered_uv_data/'

grid = xr.open_dataset('/home1/datawork/xyu/OSMOSIS_llc4320/data_rechunk/Grid.nc', decode_coords=True)
ds_U_lowpass = xr.open_zarr(out_dir_zarr+'U_lowpass_30h_filter_rechunk.zarr')
ds_V_lowpass = xr.open_zarr(out_dir_zarr+'V_lowpass_30h_filter_rechunk.zarr')

ds_filter = xr.merge([ds_U_lowpass,ds_V_lowpass,grid])
print(ds_filter)
print('\n data size: %.1f GB' %(ds_filter.nbytes / 1e9))
```

```python
ds_filter = generate_grid_ds(ds_filter, {'Z':'depth'})
gridx = xgcm.Grid(ds_filter, coords={'X':{'center':'lon', 'left':'lon_g'},
                                     'Y':{'center':'lat', 'left':'lat_g'},
                                     'Z':{'center':'depth', 'left':'depth_left'}}, periodic=['X', 'Y'])
print(gridx)
```

```python
U_low_filter = ds_filter.U_low_filter.isel(depth=16) - ds_filter.U_low_filter.isel(depth=16).mean(['time'])
V_low_filter = ds_filter.V_low_filter.isel(depth=16) - ds_filter.V_low_filter.isel(depth=16).mean(['time'])
U_low_filter
```

```python
EKE_low_filter = (gridx.interp(U_low_filter**2,'X') + gridx.interp(V_low_filter**2,'Y'))/2 # KE
EKE_low_filter
```

```python
font_size = 18

fig = plt.figure(figsize=(24,12))

ax = plt.subplot(211)
EKE.sel(lat=48.6875,lon=-16.1875,method='nearest').plot()
EKE_low_filter.sel(lat=48.6875,lon=-16.1875,method='nearest').plot()
ax.set_ylabel('Depth (m)',fontsize=font_size)
ax.set_xlabel('Time',fontsize=font_size)
ax.tick_params(labelsize=font_size)
ax.grid()

ax = plt.subplot(212)
EKE.mean(['lat','lon']).plot()
EKE_low_filter.mean(['lat','lon']).plot()
ax.set_ylabel('Depth (m)',fontsize=font_size)
ax.set_xlabel('Time',fontsize=font_size)
ax.tick_params(labelsize=font_size)
ax.grid()
```
