```python
out_dir_work = '/home/c11-data/Test_aponte/xiaolong/Parameter/data_rechunk/'

grid = xr.open_dataset(out_dir_work+'Grid.nc', decode_coords=True)
ds_U = xr.open_zarr(out_dir_work+'U_total_rechunk.zarr')
ds_V = xr.open_zarr(out_dir_work+'V_total_rechunk.zarr')
ds_B = xr.open_zarr(out_dir_work+'Buoy_total_rechunk.zarr')

ds = xr.merge([ds_U,ds_V, ds_B, grid])
print(ds)
print('\n data size: %.1f GB' %(ds.nbytes / 1e9))
```

```python
# define (real) time
def iters_to_date(iters, delta_t=3600.):
    t0 = datetime.datetime(2011,9,13,0)    
    ltime = delta_t * (np.array(iters))
    dtime = [t0+dateutil.relativedelta.relativedelta(seconds=t) for t in ltime]    
    return dtime

time_day = iters_to_date(np.arange(9415))

ds = ds.assign_coords(time=time_day)
ds
```

# 2. Filter

## low-pass filter over 1 day

```python
out_dir_zarr = '/home/c11-data/Test_aponte/xiaolong/Parameter/data_rechunk/temporary/'

def butter_filter(x, b, a):
    return xr.apply_ufunc(partial(filtfilt, b, a),
                          x.chunk(),
                          dask='parallelized',
                          output_dtypes=[x.dtype],
                          kwargs={'axis': 0})
```

```python
lowcut = 1/30/3600

fs = 1/3600
nyq = 0.5*fs
low = lowcut/nyq

b, a = butter(4, low, 'low')
```

```python
# filter
U_filter = butter_filter(ds.U.isel(depth=slice(16,39)).sel(time=slice('2012-02-25T23:00:00','2012-04-01T23:00:00')), b, a)
U_low_filter = np.real(U_filter).rename('U_low_filter')
U_low_filter = U_low_filter.to_dataset()
chunks = {'time': 865, 'lat': 1, 'lon_g': 177, 'depth':1}
U_low_filter = U_low_filter.chunk(chunks)
print(U_low_filter)

# save
file_out = out_dir_zarr+'/U_lowpass_filter.zarr'
%time U_low_filter.to_zarr(file_out, mode='w')   
```

```python
# filter
V_filter = butter_filter(ds.V.isel(depth=slice(16,39)).sel(time=slice('2012-02-25T23:00:00','2012-04-01T23:00:00')), b, a)
V_low_filter = np.real(V_filter).rename('V_low_filter')
V_low_filter = V_low_filter.to_dataset()
print(V_low_filter)

# save
file_out = out_dir_zarr+'/V_lowpass_filter_1day_rechunk.zarr'
%time V_low_filter.to_zarr(file_out, mode='w')   
```

```python
# filter
B_filter = butter_filter(ds.Buoy.isel(depth=slice(16,39)).sel(time=slice('2012-02-25T23:00:00','2012-04-01T23:00:00')), b, a)
B_low_filter = np.real(B_filter).rename('B_low_filter')
B_low_filter = B_low_filter.to_dataset()
print(B_low_filter)

# save
file_out = out_dir_zarr+'/B_lowpass_filter_rechunk.zarr'
%time B_low_filter.to_zarr(file_out, mode='w')   
```

# 3. read filtered data

```python
out_dir_zarr = '/home/c11-data/Test_aponte/xiaolong/Parameter/data_rechunk/temporary/'

grid = xr.open_dataset('/home1/datawork/xyu/OSMOSIS_llc4320/data_rechunk/Grid.nc', decode_coords=True)

ds_U_lowpass = xr.open_zarr(out_dir_zarr+'U_lowpass_filter_1day_rechunk.zarr')
ds_V_lowpass = xr.open_zarr(out_dir_zarr+'V_lowpass_filter_1day_rechunk.zarr')
ds_B_lowpass = xr.open_zarr(out_dir_zarr+'B_lowpass_filter_rechunk.zarr')

ds_filter = xr.merge([ds_U_lowpass,ds_V_lowpass,ds_B_lowpass])
print(ds_filter)
```

```python
gridx = xgcm.Grid(ds_filter, coords={'X':{'center':'lon', 'left':'lon_g'},
                                     'Y':{'center':'lat', 'left':'lat_g'}}, periodic=['X', 'Y'])
print(gridx)
```

```python
sub = 7
db_dx_filter = gridx.diff(ds_filter['B_low_filter'],'X')/ds_filter.dxC     # db/dx
db_dx_ij_filter = gridx.interp(db_dx_filter,'X')
db_dy_filter = gridx.diff(ds_filter['B_low_filter'],'Y')/ds_filter.dyC     # db/dy
db_dy_ij_filter = gridx.interp(db_dy_filter,'Y')
lateral_buoyancy_gradient_filter = np.sqrt(db_dx_ij_filter*db_dx_ij_filter + db_dy_ij_filter*db_dy_ij_filter)

lbg_filter_sub = lateral_buoyancy_gradient_filter.isel(lat=slice(0 ,None, sub),lon=slice(0 ,None, sub))
lbg_filter_sub
```

```python
lbg_filter = lbg_filter_sub.isel(lat=slice(1,-1),lon=slice(1,-1))
chunks = {'time': 865, 'lat': 1, 'lon': 24, 'depth':1}
lbg_filter = lbg_filter.chunk(chunks)
lbg_filter = np.real(lbg_filter).rename('lbg_filter')    
print(lbg_filter)
# save
file_out = out_dir_zarr+'/lbg.nc'
%time lbg_filter.to_netcdf(file_out, mode='w')   
```

## Fs

```python
sub = 7
ds_U_lowpass_sub = ds_U_lowpass.isel(lat=slice(0 ,None, sub),lon_g=slice(0 ,None, sub))
ds_V_lowpass_sub = ds_V_lowpass.isel(lat_g=slice(0 ,None, sub),lon=slice(0 ,None, sub))
ds_B_lowpass_sub = ds_B_lowpass.isel(lat=slice(0 ,None, sub),lon=slice(0 ,None, sub))

ds_filter_sub = xr.merge([ds_U_lowpass_sub,ds_V_lowpass_sub,ds_B_lowpass_sub])
print(ds_filter_sub)
```

```python
grid_sub = xgcm.Grid(ds_filter_sub, coords={'X':{'center':'lon', 'left':'lon_g'},
                                            'Y':{'center':'lat', 'left':'lat_g'}}, periodic=['X', 'Y'])
print(grid_sub)
```

```python
db_dx_filter = grid_sub.diff(ds_filter_sub['B_low_filter'],'X')/ds_filter_sub.dxC/sub     # db/dx
db_dx_ij_filter = grid_sub.interp(db_dx_filter,'X')

db_dy_filter = grid_sub.diff(ds_filter_sub['B_low_filter'],'Y')/ds_filter_sub.dyC/sub     # db/dy
db_dy_ij_filter = grid_sub.interp(db_dy_filter,'Y')

du_dx_ij_filter = grid_sub.diff(ds_filter_sub['U_low_filter'], 'X')/grid_sub.interp(ds_filter_sub.dxG, 'Y')/sub

dU_dy_1_filter = grid_sub.diff(ds_filter_sub['U_low_filter'], 'Y')/grid_sub.interp(ds_filter_sub.dyC, 'X')/sub
dU_dy_2_filter = grid_sub.interp(dU_dy_1_filter, 'X')
du_dy_ij_filter = grid_sub.interp(dU_dy_2_filter, 'Y')

dV_dx_1_filter = grid_sub.diff(ds_filter_sub['V_low_filter'], 'X')/grid_sub.interp(ds_filter_sub.dxC, 'Y')/sub
dV_dx_2_filter = grid_sub.interp(dV_dx_1_filter, 'X')
dv_dx_ij_filter = grid_sub.interp(dV_dx_2_filter, 'Y')

dv_dy_ij_filter = grid_sub.diff(ds_filter_sub['V_low_filter'], 'Y')/grid_sub.interp(ds_filter_sub.dyG, 'X')/sub

Fs_filter = -(du_dx_ij_filter*db_dx_ij_filter*db_dx_ij_filter + dv_dx_ij_filter*db_dx_ij_filter*db_dy_ij_filter + du_dy_ij_filter*db_dx_ij_filter*db_dy_ij_filter + dv_dy_ij_filter*db_dy_ij_filter*db_dy_ij_filter)
Fs_filter
```
