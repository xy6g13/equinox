## Band-integrated energy

```python
E_Lagrangian = ds.E_Lagrangian
E_Eulerian = ds.E_Eulerian
E_Eulerian
```

```python
df=E_Eulerian.frequency[1]-E_Eulerian.frequency[0]
print(df)
```

```python
# Inertial frequency
f_llc = 2.*2.*np.pi/86400. * np.sin(np.deg2rad(E_Eulerian.lat))
f_cpd_llc = f_llc*86400/2./np.pi
f_cpd_llc.plot()
```

```python
# compare each component by applying windowing
labels = ['Total', 'Low-frequency (<0.5 cpd)','High-frequency (>0.5 cpd)','Semi-diurnal', 'Diurnal', 'Near-inertial']


E_KE_Lagrangian = (E_Lagrangian*df).sum(dim='frequency')
E_high_Lagrangian = (E_Lagrangian.where((abs(E_Lagrangian.frequency)>0.5) , other=0.)*df).sum(dim='frequency')

E_KE_Eulerian = (E_Eulerian*df).sum(dim='frequency')
E_high_Eulerian = (E_Eulerian.where((abs(E_Eulerian.frequency)>0.5) , other=0.)*df).sum(dim='frequency')


fig, ax = plt.subplots(figsize=(20,16))

ax = plt.subplot(1,2,1)
E_KE_Lagrangian.plot(color='black', label=labels[0], linewidth=2.5) # total energy
(E_KE_Lagrangian - E_high_Lagrangian).plot(color='red', label=labels[1], linewidth=2.5) # total energy
E_high_Lagrangian.plot(color='blue', label=labels[2], linewidth=2.5)   # energy at low frequencies (> 5 day
(E_Lagrangian.where((abs(E_Lagrangian.frequency)>1.9) & (abs(E_Lagrangian.frequency)<2.1) , other=0.)*df).sum(dim='frequency').plot(color='yellow', label=labels[3], linewidth=2.5) # energy around semi-diurnal frequencies
(E_Lagrangian.where((abs(E_Lagrangian.frequency)>0.9) & (abs(E_Lagrangian.frequency)<1.1) , other=0.)*df).sum(dim='frequency').plot(color='m', label=labels[4], linewidth=2.5) # energy around diurnal frequencies
(E_Lagrangian.where(abs(E_Lagrangian.frequency+f_cpd_drifter)<.1, other=0.)*df).sum(dim='frequency').where(abs(E_Lagrangian.lat)>15).plot(color='cyan', label=labels[5], linewidth=2.5) # energy around f
ax.legend(prop={'size': 20}); ax.grid()
# x,y ticks
ax.set_xlim(-60., 60.)
ax.set_xticks([-60., -40., -20., 0., 20., 40., 60.])
ax.set_xticklabels(['60S','40S','20S','0N','20N','40N','60N'])
#ax.set_ylim([0,0.4])
ax.set_yscale('log')
ax.set_ylim([1e-4,1e1])
# Label, size
ax.set_xlabel('Latitude',fontsize=24)
ax.set_ylabel('Kinetic energy ($u^2 + v^2, m^2 s^{-2}$)',fontsize=24)
ax.tick_params(labelsize=24)
ax.set_title('LLC4320 (Lagrangian)',fontsize=24)

ax = plt.subplot(1,2,2)
E_KE_Eulerian.plot(color='black', label=labels[0], linewidth=2.5) # total energy
(E_KE_Eulerian - E_high_Eulerian).plot(color='red', label=labels[1], linewidth=2.5) # total energy
E_high_Eulerian.plot(color='blue', label=labels[2], linewidth=2.5)   # energy at low frequencies (> 5 day
(E_Eulerian.where((abs(E_Eulerian.frequency)>1.9) & (abs(E_Eulerian.frequency)<2.1) , other=0.)*df).sum(dim='frequency').plot(color='m', label=labels[3], linewidth=2.5) # energy around semi-diurnal frequencies
(E_Eulerian.where((abs(E_Eulerian.frequency)>0.9) & (abs(E_Eulerian.frequency)<1.1) , other=0.)*df).sum(dim='frequency').plot(color='green', label=labels[4], linewidth=2.5) # energy around diurnal frequencies
(E_Eulerian.where(abs(E_Eulerian.frequency+f_cpd_drifter)<.1, other=0.)*df).sum(dim='frequency').where(abs(E_Eulerian.lat)>15).plot(color='cyan', label=labels[5], linewidth=2.5) # energy around f
ax.legend(prop={'size': 20}); ax.grid()
# x,y ticks
ax.set_xlim(-60., 60.)
ax.set_xticks([-60., -40., -20., 0., 20., 40., 60.])
ax.set_xticklabels(['60S','40S','20S','0N','20N','40N','60N'])
#ax.set_ylim([0,0.4])
ax.set_yscale('log')
ax.set_ylim([1e-4,1e1])
# Label, size
ax.set_xlabel('Latitude',fontsize=24)
ax.set_ylabel('Kinetic energy ($u^2 + v^2, m^2 s^{-2}$)',fontsize=24)
ax.tick_params(labelsize=24)
ax.set_title('LLC4320 (Eulerian)',fontsize=24)
```
